```python
import pandas as pd
import numpy as np
import os 
import csv
```

```python
csv_1 = "Data/JC-201804-citibike-tripdata.csv"
csv_2 = "Data/JC-201805-citibike-tripdata.csv"

April_data = pd.read_csv(csv_1, encoding="utf-8")
May_data = pd.read_csv(csv_2, encoding="utf-8")
```

```python
April_data.head()
```

```python
May_data.head()
```

```python
all_data = pd.concat([April, May])
all_data.head()
```

```python
all_data['short-term customer'] = (all_data['usertype'] == 'Customer')
all_data['long-term customer'] = (all_data['usertype'] == 'Subscriber')
all_data.head()
```

```python
all_data['gender'].replace(0, 'Female',inplace=True)
all_data['gender'].replace(1, 'Male',inplace=True)
all_data['male'] = [all_data.loc['gender'] == 'male']
all_data['female'] = [all_data.loc['gender'] == 'female']
all_data.head()
```

```python
all_data['short-term customer'] = all_data['short-term customer'].astype(int)
all_data['long-term customer'] = all_data['long-term customer'].astype(int)
all_data.head()

```

```python
all_data.to_csv('cleaned_data.csv', encoding='utf-8', index=True)
```
